```python
import os
import glob

def renombrar_imagenes(ruta):
    # Verificar si la ruta existe y es un directorio
    if not os.path.exists(ruta):
        print(f"La ruta {ruta} no existe.")
        return
    
    if not os.path.isdir(ruta):
        print(f"La ruta {ruta} no es un directorio.")
        return

    # Buscar todas las imgenes en la ruta (considerando las extensiones ms comunes)
    extensiones = ["*.png", "*.jpg", "*.jpeg", "*.bmp", "*.gif"]
    imagenes = []
    for extension in extensiones:
        imagenes.extend(glob.glob(os.path.join(ruta, extension)))

    if not imagenes:
        print(f"No se encontraron imgenes en la ruta {ruta}.")
        return

    # Renombrar cada imagen
    for i, imagen in enumerate(imagenes):
        try:
            # Obtener la extensin del archivo
            extension = os.path.splitext(imagen)[1]
            # Crear el nuevo nombre
            nuevo_nombre = f"tornado{i}{extension}"  # Aqu colocar el nombre deseado de las imgenes
            # Obtener la ruta completa del nuevo nombre
            nueva_ruta = os.path.join(ruta, nuevo_nombre)
            # Renombrar el archivo
            os.rename(imagen, nueva_ruta)
            print(f"Renombrado: {imagen} -> {nueva_ruta}")
        except Exception as e:
            print(f"No se pudo renombrar {imagen} -> {nueva_ruta}. Error: {e}")

    print(f"Renombrado completado. Total de imgenes renombradas: {len(imagenes)}")

# Forma de usar
ruta_imagenes = r"D:\Escuela Betin\9no Sem\3. Inteligencia Artificial\CNN\IMG\tornado"
renombrar_imagenes(ruta_imagenes)
```
